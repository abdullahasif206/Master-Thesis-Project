```python
import numpy as np
import pandas as pd

rotation_angle_x = np.deg2rad(5.553046427308958)  
rotation_angle_y = np.deg2rad(0)      
rotation_angle_z = np.deg2rad(-0.5625267599712734)  

rotation_matrix_x = np.array([
    [1, 0, 0],
    [0, np.cos(rotation_angle_x), np.sin(rotation_angle_x)],
    [0, -np.sin(rotation_angle_x), np.cos(rotation_angle_x)]
])

rotation_matrix_y = np.array([
    [np.cos(rotation_angle_y), 0, -np.sin(rotation_angle_y)],
    [0, 1, 0],
    [np.sin(rotation_angle_y), 0, np.cos(rotation_angle_y)]
])

rotation_matrix_z = np.array([
    [np.cos(rotation_angle_z), np.sin(rotation_angle_z), 0],
    [-np.sin(rotation_angle_z), np.cos(rotation_angle_z), 0],
    [0, 0, 1]
])

# Calculate the combined rotation matrix
r_combined = np.matmul(np.matmul(rotation_matrix_x, rotation_matrix_y), rotation_matrix_z)

# Define the path to your accelerometer data file
data_file_path = r"C:\Users\uig38452\Master Thesis\Data Recording\all_data_LSB_Initial\WLU_WLD\WLU_WLD_Max.csv"
# Read accelerometer and gyroscope data
data = pd.read_csv(data_file_path)

# Create an empty list to store transformed coordinates
transformed_data = []

# Apply the 3D rotation matrix to sensor data
for index, row in data.iterrows():
    ax, ay, az, gx, gy, gz = row['AX_LSB'], row['AY_LSB'], row['AZ_LSB'], row['GX_LSB'], row['GY_LSB'], row['GZ_LSB']
    
    # Transform accelerometer data
    transformed_accelerometer_data = np.dot(r_combined, [ax, ay, az])/16
    
    # Transform gyroscope data
    transformed_gyroscope_data = np.dot(r_combined, [gx, gy, gz])/16
    
    # Append transformed data to the list
    transformed_data.append([*transformed_accelerometer_data, *transformed_gyroscope_data])

# Create a DataFrame with all transformed data
transformed_data_df = pd.DataFrame(transformed_data, columns=['AX_LSB', 'AY_LSB', 'AZ_LSB', 'GX_LSB', 'GY_LSB', 'GZ_LSB'])

# Define the path to save the combined transformed data to a CSV file
output_file_path = 't_WLU_WLD_Old_Koordinate_Transform_Max.csv'

# Save the DataFrame to a CSV file
transformed_data_df.to_csv(output_file_path, index=False)

print(f"Transformed data saved to {output_file_path}")
```
